```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
raw_data_dir= './data/'
# 33 families of products
# index is family,store_nbr
train_df = pd.read_csv(raw_data_dir+'kag-store-sales-train.csv')

train_df\
    .loc[lambda x:x['family']=='AUTOMOTIVE']\
        .loc[lambda x:x['store_nbr']==10]\
    .set_index('date')['sales'].plot()
train_df.groupby('date').sum()[['sales']].plot()
```

```python
train_df\
    .loc[lambda x:x['family']=='LIQUOR,WINE,BEER']\
    .groupby('date').sum()[['sales']].ewm(halflife=5).mean().plot()
```
