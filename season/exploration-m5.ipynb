```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from crptmidfreq.season.m5_data import get_data
```

```python
df=get_data()
```

```python
df['dscode_str'] = df['dept_id']+'_'+df['store_id']
```

```python
dscodes=df['dscode_str'].unique()
for dscode in dscodes:
    dfloc=df[df['dscode_str']==dscode]
    dfloc.set_index('date')[['sales_log']].plot(title=dscode)
    plt.figure()
```
