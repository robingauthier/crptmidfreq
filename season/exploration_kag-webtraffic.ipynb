```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
raw_data_dir= './data/'
# 33 families of products
# index is family,store_nbr
train_df = pd.read_csv(raw_data_dir+'kag-webtraffic-train_2.csv')
```

```python

train_pdf = pd.melt(train_df, 
                  id_vars = ['Page'], 
                  var_name = 'date', 
                  value_name = 'traffic')
```

```python
train_pdf.loc[lambda x:x['Page']=='2NE1_zh.wikipedia.org_all-access_spider']\
    .set_index('date')[['traffic']]\
    .plot()
```

```python
train_pdf.groupby('date')[['traffic']].sum().plot()
```
